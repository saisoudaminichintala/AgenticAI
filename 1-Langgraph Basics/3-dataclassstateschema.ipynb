### Dataclasses

- A dataclass is a lightweight Python class designed primarily for storing data. 
- The @dataclass decorator automatically generates methods such as __init__, __repr__, and __eq__, reducing boilerplate code.
-  Dataclasses are commonly used for structured internal data objects, whereas Pydantic models are preferred when runtime validation and schema generation are required.

```python
from dataclasses import dataclass
from typing import Literal

@dataclass
class DataClassState:
    name:str
    game:Literal["Basketball", "Soccer"]
```

```python
def start_play(state: DataClassState):
    print("start_play node has been called.")
    return{"graph_info": state.name+ "I'm starting to play. "}


def play_Soccer(state: DataClassState):
    print("play_Soccer node has been called.")
    return{"graph_info": state.name+ "Soccer is fun! "}

def play_Basketball(state: DataClassState):
    print("play_Basketball node has been called.")
    return{"graph_info": state.name+ "Basketball is fun! "}
```

```python
import random
from typing import Literal
def random_choice(state: DataClassState)->Literal["play_Soccer", "play_Basketball"]:
    name =state.name
    
    if random.random() < 0.5:
        return "play_Soccer"
    else:
        return "play_Basketball"
```

```python
from IPython.display import Image, display

from langgraph.graph import StateGraph, START, END

# Build the graph
graph = StateGraph(DataClassState)

# Add nodes
graph.add_node("start_play", start_play)
graph.add_node("play_Soccer", play_Soccer)
graph.add_node("play_Basketball", play_Basketball)

# Add edges
graph.add_edge(START, "start_play")
graph.add_conditional_edges("start_play", random_choice)
graph.add_edge("play_Soccer", END)
graph.add_edge("play_Basketball", END)

# Compile the graph
graph_builder = graph.compile()

# View the graph
display(Image(graph_builder.get_graph().draw_mermaid_png()))
```

```python
graph_builder.invoke(DataClassState(name="John", game="Basketball"))
```
