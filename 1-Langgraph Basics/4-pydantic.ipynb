### Pydantic Data Validation

- A Python data validation library that uses type annotations to define schemas.
-  It provides runtime validation, automatic type conversion, serialization, and JSON schema generation.
- Commonly used for tool input schemas, structured LLM outputs, API request validation

```python
from langgraph.graph import StateGraph, START, END
from pydantic import BaseModel

class PydanticState(BaseModel):
    name: str
    
```

```python
#Node functions
def start_play(state: PydanticState):
    print("start_play node has been called.")
    return{"name": state.name+ " I'm starting to play. "}

def play_Soccer(state: PydanticState):
    print("play_Soccer node has been called.")
    return{"name": state.name+ "Soccer is fun! "}

def play_Basketball(state: PydanticState):
    print("play_Basketball node has been called.")
    return{"name": state.name+ "Basketball is fun! "} 

import random
def random_choice(state: PydanticState)->str:
    name =state.name
    
    if random.random() < 0.5:
        return "play_Soccer"
    else:
        return "play_Basketball"
```

```python
# Graph construction
graph = StateGraph(PydanticState)
graph.add_node("start_play", start_play)
graph.add_node("play_Soccer", play_Soccer)
graph.add_node("play_Basketball", play_Basketball)

# Add edges

graph.add_edge(START, "start_play")

graph.add_conditional_edges(
    "start_play",
    random_choice,
    {
        "play_Soccer": "play_Soccer",
        "play_Basketball": "play_Basketball",
    }
)

graph.add_edge("play_Soccer", END)
graph.add_edge("play_Basketball", END)




graph_builder = graph.compile()

# Invoke the graph
graph_builder.invoke({"name": "John"})
```
